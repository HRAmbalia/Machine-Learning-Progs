````python
import numpy as np

# Set a specific style for the plot
plt.style.use('_mpl-gallery')

# Generate reproducible random data using NumPy
x = np.linspace(0, 8, 16) # will generate an array of 16 equally spaced values that start from 0 and end at 8.
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))  # to generate array of random numbers that are uniformly distributed between 0.0 and 0.5.
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))
# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Fill the area between y1 and y2 with a translucent fill color
ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)

# Plot the average of y1 and y2 as a line
ax.plot(x, (y1 + y2)/2, linewidth=2)

# Set x and y axis limits and ticks
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

# Display the plot
plt.show()
````
