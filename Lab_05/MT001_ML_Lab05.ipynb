```python
# Importing library
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# Importing the dataset
import pandas as pd
file_path = "/content/gdrive/MyDrive/Colab Notebooks/ML_Lab05_weather_data.csv"
weather_data = pd.read_csv(file_path)
print(weather_data.columns)
```

```python
# Performing data preprocessing

# Dropping the unnecessary columns
columns_to_drop = ['Formatted Date', 'Loud Cover']
weather_data = weather_data.drop(columns=columns_to_drop)
print(f'Weather_data columns : {weather_data.columns}')

# Label Encoding
from sklearn.preprocessing import LabelEncoder
def perform_encoding(column_name):
    label_encoder = LabelEncoder()
    encoded_column_name = label_encoder.fit_transform(column_name)
    return encoded_column_name
weather_data['Summary'] = perform_encoding(weather_data['Summary'])
weather_data['Precip Type'] = perform_encoding(weather_data['Precip Type'])
weather_data['Daily Summary'] = perform_encoding(weather_data['Daily Summary'])

# Taking Summary as the lebel comumn
weather_labels = weather_data['Summary']
weather_data = weather_data.drop('Summary', axis=1)

# Dropping the highly corelated columns
import numpy as np
corr = weather_data.corr()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] >= 0.8)]
print(f'Highly corelated columns(will be dropped) : {to_drop}')
weather_data = weather_data.drop(to_drop, axis=1)

print("Final dataset")
print(weather_data)
```

```python
# Performing data preprocessing

# Performing scaling
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()

rows = ['Temperature (C)', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)', 'Pressure (millibars)', 'Daily Summary']
header_values = weather_data.columns
print(header_values)
for column in rows:
    if column in header_values:
        scaler.fit(weather_data[rows])
        weather_data[rows] = scaler.transform(weather_data[rows])

print("Final dataset after scaling")
print(weather_data)
```

```python
# Split the data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(weather_data, weather_labels, test_size=0.05, random_state=42) # 95%-05%
```

```python
# Training the model
from sklearn.tree import DecisionTreeClassifier
weather_model = DecisionTreeClassifier(max_leaf_nodes=2)
weather_model.fit(X_train, y_train)
```

```python
y_pred = weather_model.predict(X_test)
accuracy = np.mean(y_pred == y_test)
print('Accuracy on test set:', accuracy)
```

```python
# Finding precision, Recall and accuracy
from sklearn.metrics import accuracy_score, precision_score, recall_score
y_pred = weather_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
print(f'Accuracy : {accuracy}')
print(f'Precision : {precision}')
print(f'Recall : {recall}')
```

```python
# visualizing
import matplotlib.pyplot as plt
import six
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

```python
# Function to plot the decision boundry BUT IT'S NOT PRINTING ANY BOUNDARY.!!! WHY???
from matplotlib.colors import ListedColormap
def plot_decision_boundary(clf, X, y, legend=False, plot_training=True, ax=None, figsize=(10,6)):

    if not ax:
        f, ax = plt.subplots(figsize=figsize)

    # Generate a grid of points with 30 features
    x1s = np.linspace(np.min(X[:, 0]) - 0.1 * np.mean(X[:, 0]), 1.1 * np.max(X[:, 0]), 100)
    x2s = np.linspace(np.min(X[:, 1]) - 0.1 * np.mean(X[:, 1]), 1.1 * np.max(X[:, 1]), 100)
    x1, x2 = np.meshgrid(x1s, x2s)

    # Create an array with 30 columns, each representing a feature
    # Here, we are assuming that X_train has 30 features
    X_new = np.column_stack((x1.ravel(), x2.ravel(), np.zeros((x1.ravel().shape[0], 28))))

    print("X_new shape:", X_new.shape)
    y_pred = clf.predict(X_new).reshape(x1.shape)

    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    ax.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)

    custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
    ax.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)

    dots = ["yo", "bs", "g^"]

    if plot_training:
        for i in np.unique(y):

            ax.plot(X[:, 0][y==i], X[:, 1][y==i], dots[i], label=str(i))

    else:
        ax.set_xlabel(r"$x_1$", fontsize=18)
        ax.set_ylabel(r"$x_2$", fontsize=18, rotation=0)
    ax.legend(loc="lower right", fontsize=14)
    ax.set_title("Decision boundary")

    return ax
```

```python
# Pinting the tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(breast_cancer_model,filled=True)
```

```python
# printing all the alpha values
path = breast_cancer_model.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
print(ccp_alphas)
print(impurities)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': ccp_alphas}
ridge = Ridge()
grid_search = GridSearchCV(ridge, param_grid, cv=5)
grid_search.fit(X_train, y_train)
best_alpha = grid_search.best_params_['alpha']
print(f'Best alpha value is : {best_alpha}')
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    breast_cancer_model = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    breast_cancer_model.fit(X_train, y_train)
    clfs.append(breast_cancer_model)
print("Number of nodes in the last tree is: {} with ccp_alpha: {}".format(
      clfs[-1].tree_.node_count, ccp_alphas[-1]))
```

```python
train_scores = [breast_cancer_model.score(X_train, y_train) for breast_cancer_model in clfs]
test_scores = [breast_cancer_model.score(X_test, y_test) for breast_cancer_model in clfs]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```
