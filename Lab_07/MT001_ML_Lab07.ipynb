```python
# Adding names to the columns
import pandas as pd
column_names = ['label', 'id', 'time', 'extra', 'name', 'text']
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Lab07_dataset.csv', encoding='latin-1', names=column_names)
df.head()
```

```python
# Dropping extra columns
columns_to_drop = ['id', 'time', 'extra', 'name']
df.drop(columns=columns_to_drop, inplace=True)
df.head()
```

```python
df = df[df['label'].isin([0, 4])].groupby('label').head(5000)
df['label'] = df['label'].replace(4,1)
df
```

```python
positive_string_array = df[df['label'] == 1]['text'].tolist()
negative_string_array = df[df['label'] == 0]['text'].tolist()
```

```python
import re
def remove_links(sentence):
    return re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', sentence)

positive_without_link = [remove_links(sentence) for sentence in positive_string_array]
negative_without_link = [remove_links(sentence) for sentence in negative_string_array]
```

```python
def remove_at_words(sentence):
    return re.sub(r'@\w+\s?', '', sentence)

positive_without_at_words = [remove_at_words(sentence) for sentence in positive_without_link]
negative_without_at_words = [remove_at_words(sentence) for sentence in negative_without_link]
```

```python
import re
def remove_punctuation(list_of_string):
  punctuation_marks = [',', '.', '(', ')', '!', '@', '#', '$', '%', '^', '&', '*', '<', '>', '?', ';', ':', '-', '_', '+', '-', '"', "'"]
  list_of_strings = []
  for sublist in list_of_string:
    if isinstance(sublist, str):
      for punctuation_mark in punctuation_marks:
        sublist = sublist.replace(punctuation_mark, '')
        sublist = re.sub(r'\d+', '', sublist)
      list_of_strings.append(sublist)
  return list_of_strings

positive_without_punctuation = remove_punctuation(positive_without_at_words)
negative_without_punctuation = remove_punctuation(negative_without_at_words)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords_english = stopwords.words('english')
print(stopwords_english)
def remove_stop_words(sentence):
    words = sentence.split()
    filtered_words = [word for word in words if word.lower() not in stopwords_english]
    return ' '.join(filtered_words)

positive_without_stop_words = [remove_stop_words(sentence) for sentence in positive_without_punctuation]
negative_without_stop_words = [remove_stop_words(sentence) for sentence in negative_without_punctuation]
```

```python
document = positive_without_stop_words + negative_without_stop_words
sentiment = [1] * len(positive_without_stop_words) + [0] * len(negative_without_stop_words)
```

Count Vectorize

```python
from sklearn.feature_extraction.text import CountVectorizer
count_matrix = CountVectorizer()
count_matrix_final = count_matrix.fit_transform(document)
newarray = count_matrix_final.toarray()
```

```python
count_matrix.vocabulary_
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                                                count_matrix_final,
                                                sentiment,
                                                test_size=0.2,
                                                random_state=0)
```

```python
test = X_test.toarray()
```

Classifier 01

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(X_train, y_train)
```

```python
pred = classifier.predict(X_test)
pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

```python
classifier.predict_proba(X_test)
```

```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
weather = ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast', 'Overcast', 'Rainy']
temp = ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild']
play=['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
```

```python
le = preprocessing.LabelEncoder()
weather_encoded=le.fit_transform(weather)
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)
print("Temp:",temp_encoded)
print("Play:",label)
print("Weather:" ,weather_encoded)
```

```python
features=tuple(zip(weather_encoded,temp_encoded))
print("Features:",features)
```

```python
model=MultinomialNB(alpha=0, fit_prior=True)
#model=MultinomialNB(alpha=0, force_alpha=True)
model.fit(features,label)
model.get_params()
```

```python
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
print("Predicted Value:", predicted)
```

```python
predicted= model.predict([[0,1]]) # 0:Overcast, 1:Hot
print("Predicted Value:", predicted)
```

```python
predicted= model.predict([[2,2]]) # 2:Sunny, 2:Mild
print("Predicted Value:", predicted)
```
